```python
%matplotlib inline 
import matplotlib.pyplot as plt
import seaborn
seaborn.set()
from numpy import exp, array, append
```

```python
#sigmoid activation function

def Sigmoid(x):
    return 1 / (1 + exp(-x))
```

```python
#Graphical Representation of sigmoid activation function

for i in array([range(-10, 10)]):
    x = append(array([]), i )
    plt.plot(x, Sigmoid(x))
    plt.title("Sigmoid activation function")
    plt.xlabel("x")
    plt.ylabel("S(x)")
```

```python
# Tan-h activation function

def Tanh(x):
    return 2 / (1 + exp(-2 * x )) - 1
```

```python
#Graphical represenation of our tan-h activation function

for i in array([range(-10,10)]):
    x = append(array([]), i)
    plt.plot(x, Tanh(x))
    plt.title("Tan-h activation function")
    plt.xlabel("x")
    plt.ylabel("tanh(x)")
```

```python
from numpy import sum

#Softmax activation function 

def Softmax(x):
    return exp(x) / sum(exp(x))

x = array([1,2,3,4,5])

Softmax(x)
```

```python
#ReLU activation function

def ReLU(x):
    return x * (x > 0)

# Graphical representation of ReLU activation

x = array(range(-10, 10))
plt.plot(x,ReLU(x))
plt.title("ReLU activation function")
plt.xlabel("x")
plt.ylabel("ReLU(x)")
```

```python
#Leaky ReLU

from numpy import where

#Leaky ReLU activation function

def LeakyReLU(x):
    return where(x>0, x, x * 0.1)

#Graphical representation of our Leaky ReLU activation function 

x = array(range(-10, 10))
plt.plot(x, LeakyReLU(x))
plt.title("Leaky ReLU activation function")
plt.xlabel("x")
plt.ylabel("Leaky ReLU(x)")
```
