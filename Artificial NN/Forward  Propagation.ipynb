```python
from numpy import dot
from numpy import exp, array, append

def Sigmoid(x):
    return 1 / (1 + exp(-x))

def ReLU(x):
    return x * (x > 0)
```

```python

i = array([0.1,0.4,0.5])

w_ij = array([[0.1, 0.2, 0.4, 0.3],
              [0.5, 0.4, 0.7, 0.9],
              [0.2, 0.6, 0.3, 0.8]
             ])

w_jk = array([[0.2],
              [0.3],
              [0.6],
              [0.1]
             ])

#dot product of input and hidden layer
z_ij = dot(i, w_ij) 
print(z_ij)

#applying  the ReLU activation function
a_ij = ReLU(z_ij)
print(a_ij)

#dot product of inputs in hidden layer after activation
#and weights of output layer
z_jk = dot(a_ij, w_jk)

#applying Sigmoid activation function 
a_jk = Sigmoid(z_jk)

#The final output of feed forward neural network
a_jk
```

```python
#cross entropy

from numpy import log10
# cross entropy error function
def crossentropyerror(a, y):
    return - sum(y * log10(a) + (1 - y) * log10(1 - a))
a = a_jk # the output of the neural network
y = array([[1]]) # the ground truth
crossentropyerror(a, y)
# the cross entropy error equals 0.19513122954797735
```

```python
#Back Propagation 
# Derivative of the Sigmoid activation function
def SigmoidDerivative(x):
    return Sigmoid(x) * (1 - Sigmoid(x))
# Graphical representation of the Sigmoid derivative
x = array(range(-10, 10))
plt.plot(x, SigmoidDerivative(x))
plt.title("Derivative of Sigmoid")
plt.xlabel("x")
plt.ylabel("Sigmoid Derivative")
```

```python
# Derivative of the Sigmoid activation function
def ReLUDerivative(x):
    return 1. * (x > 0)
# Graphical representation of the ReLU derivative
x = array(range(-10, 10))
plt.plot(x, ReLUDerivative(x))
plt.title("Derivative of ReLU")
plt.xlabel("x")
plt.ylabel("ReLU Derivative")
```
