```python
import numpy as np
import os
from scipy.misc import imread, imresize
import matplotlib.pyplot as plt
%matplotlib inline

print('packagees loaded')
```

```python
cwd = os.getcwd()
print("current folder is %s" %(cwd))
```

```python
def print_typeshape(img):
    print('Type is %s' %(type(img)))
    print(' Shape is %s'%(img.shape,))
```

```python
iron_man = imread('./data/ironman.jpeg')
print_typeshape(iron_man)
```

```python
plt.figure(0)
plt.imshow(iron_man)
plt.title('Iron man')
plt.draw()
```

```python
#load + cast to float

iron_man2 = imread('./data/ironman.jpeg').astype(np.float)
print_typeshape(iron_man2)

plt.figure(0)
plt.imshow(iron_man2)
plt.title('Original image with imread.astype(np.float)')
plt.draw()
```

```python
#load + cast to float

iron_man3 = imread('./data/ironman.jpeg').astype(np.float)
print_typeshape(iron_man3)

plt.figure(0)
plt.imshow(iron_man3/255.)
plt.title('Original image with imread.astype(np.float)/255')
plt.draw()
```

```python
#resize 

ironman = imresize(iron_man, [100,100,3])
print_typeshape(iron_man)

plt.figure(1)
plt.imshow(ironman)
plt.title('Resized image')
plt.draw()
```

```python
def rgb2gray(rgb):
    if len(rgb.shape) is 3:
        return np.dot(rgb[...,:3], [0.299, 0.587, 0.114])
    else:
        print ("Current Image if GRAY!")
        return rgb
```

```python
iron_man_smallgray = rgb2gray(iron_man)
print ('size of iron_man_smallgray is %s' %(iron_man_smallgray.shape,))
print ('type of iron_man_smallgray is ', type(iron_man_smallgray))
```

```python
plt.imshow(iron_man_smallgray, cmap=plt.get_cmap('gray'))
plt.title('[imshow] Gray image')
plt.colorbar()
plt.draw()
```

```python
#reshape 

iron_man_rowvec  = np.reshape(iron_man_smallgray, (1,-1))
print ('size of iron_man_rowvec is %s' % (iron_man_rowvec.shape,))
print (' type of iron_man_rowvec is ', type(iron_man_rowvec))

#vector to matrix

iron_man_matrix = np.reshape(iron_man_rowvec, (340, -1))
print ( 'size of iron_man_matrix is %s' %(iron_man_matrix.shape,))
print ("type of iron_man_matrix is", type(iron_man_matrix))
```

```python
plt.imshow(iron_man_matrix, cmap=plt.get_cmap("gray"))
plt.title("Reshaped Matrix")
plt.colorbar()
plt.draw()
```

```python
cwd = os.getcwd()
path  = cwd + '/data/harry'
flist = os.listdir(path)
print ("[%d] FILE ARE IN [%s]" % (len(flist), path))
```

```python
for i, f in enumerate(flist):
    print ("[%d]TH FILE IS [%s]" % (i, f))
```

```python
valid_exts = [".jpg",".gif",".png",".tga", ".jpeg"]
imgs = []
names = []
for f in flist:
    ext = os.path.splitext(f)[1]
    if ext.lower() not in valid_exts:
        continue
    # LOAD AND APPEND TO IMGS LIST
    fullpath = os.path.join(path,f)
    imgs.append(imread(fullpath))
    # APPEND NAMES
    names.append(os.path.splitext(f)[0])
```

```python
for img, name in zip(imgs, names):
    plt.imshow(img)
    plt.title(name)
    plt.show()
```
